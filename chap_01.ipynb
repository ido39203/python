```python
# 변수(Variable) 선언 연습
radius = 100 # 변수는 =의 왼쪽에 제일 먼저 나타난다. 100에 radinus에 저장.
# 변수 이름 앞 뒤 ""를 사용하지 않는다.
print(radius) # 변수 radius에 저장된 값 출력
```

```python
area = 1000*1000*pi # pi는 변수로서 정의(define, =왼쪽에 먼저 나타나야 함)되지 않고 사용되었으므로 오류.
```

```python
pi = 3.14 # pi가 3.14라는 값으로 정의됨.
area = 100*100*pi # 이미 정의된 pi라는 변수를 =오른쪽에 사용 가능함.
print(pi)
print(area)
pi = 3.141519 # pi를 재정의함. 변수는 값을 변경할 수 있음.
print(pi)
```

```python
# 한 줄에 여러 값 출력 연습
print("pi = ", pi, "면적", area) # ,로 한 줄에 출력할 여러 값을 써 줄 수 있다.
```
